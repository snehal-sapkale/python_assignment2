#### 1) Use the given file named cricket.csv and perform the following operations, Read the file in DataFrame

```python
import pandas as pd
import os
```

```python
path = os.getcwd() + "\datasets\\"

df = pd.read_csv(path + "cricket.csv")
df
```

#### 2) List the name of cricketer and their respective runs

```python
#df[['name', 'runs']]
df.loc[:, ['name','runs']]
```

#### 3) Find total wickets taken by them

```python
df.loc[:, ['name','runs','wickets']]
```

#### 4) Find average of catches taken

```python
df['catches'].mean()
```

#### 5) Find the name of wicketkeeper

```python
x = df[(df['stumpings'] > 0 )] 
x.name
```

#### 6) print the name of bowler who played highest number of matches

```python
max_matches = df['matches'].max()
x = df[(df['matches'] == max_matches)] 
x.name
```

#### 7) Find average of all the bowlers

```python
x = df[df['wickets'] != 0]
df['bowl_average'] = x['runs'] / x['wickets'] 
df
```

#### 8) Find name of the bowler with least bowling average (wickets per match).

```python
least_average = df['bowl_average'].min()
x = df[(df['bowl_average'] == least_average)] 
x.name
```

#### 9) Print information about all players whose catches per match ratio is greater than 0.5 store it in a csv file

```python
df['catch_per_match'] = x['catches'] / x['matches'] 
x = df[(df['catch_per_match'] > 0.5)]
x
x.to_csv('output.csv', index = False)
```
