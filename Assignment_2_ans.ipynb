#### 1 - 1) Use the given file named cricket.csv and perform the following operations, Read the file in DataFrame

#### 2) List the name of cricketer and their respective runs

```python
df['catches'].mean()
```

#### 5) Find the name of wicketkeeper

```python
x = df[(df['stumpings'] > 0 )] 
x.name
```

#### 6) print the name of bowler who played highest number of matches

```python
max_matches = df['matches'].max()
x = df[(df['matches'] == max_matches)] 
x.name
```

#### 7) Find average of all the bowlers

```python
x = df[df['wickets'] != 0]
df['bowl_average'] = x['runs'] / x['wickets'] 
df
```

#### 8) Find name of the bowler with least bowling average (wickets per match).

```python
least_average = df['bowl_average'].min()
x = df[(df['bowl_average'] == least_average)] 
x.name
```

#### 9) Print information about all players whose catches per match ratio is greater than 0.5 store it in a csv file

```python
df['catch_per_match'] = x['catches'] / x['matches'] 
x = df[(df['catch_per_match'] > 0.5)]
x
x.to_csv('output.csv', index = False)
```

#### 6) In which hour of the day maximum posts are posted?

```python
df = pd.read_csv(path + "fb.csv")
x = df['Post Hour'].value_counts().rename_axis('hour').to_frame('max_count')
print("Hour of the day that has max posts is: ", x.iloc[[0], :1])
```

#### 7) In which day of October, maximum number of interactions are found?

```python
df = pd.read_csv(path + "fb.csv")
x = df[df['Post Month'] == 10]
x1 = x[['Post Weekday','Total Interactions']]
x2 = x1[x1['Total Interactions'] == x1['Total Interactions'].max()]

print("Maximum number of interactions are found on: ", x2.iloc[[0], :1])
```

#### 9) Which kind of posts are posted maximum number of times in November month?

```python
df = pd.read_csv(path + "fb.csv")
x = df[df['Post Month'] == 11]
x1 = x['Type'].value_counts().rename_axis('Post Type').to_frame('max_count')
x2 = x1[x1['max_count'] == x1['max_count'].max()]
print("Posts type that are posted maximum number of times in November month: ", x2.iloc[[0], :1])
```

#### 10) Find average and median of interactions for the paid (1) posts

```python
df = pd.read_csv(path + "fb.csv")
x = df[df['Paid'] == 1]
average = x['Total Interactions'].mean()
median = x['Total Interactions'].median()

print("average: ", average)

print("median: ", median)
```
